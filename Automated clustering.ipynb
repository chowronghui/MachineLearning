```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import matplotlib.cm as cm
```

```python
df = pd.read_csv("9thDecember2017_31July2018_1.csv")
y = df['Value']
x = df['Date']
label = df['Label']
```

```python
#https://stackoverflow.com/questions/12236566/setting-different-color-for-each-series-in-scatter-plot-on-matplotlib
```

```python
y_min = min(y)
y_max = max(y)
print(y_min)
print(y_max)
first_range = y_min + ((y_max-y_min)/5)
print(first_range)
```

```python
np.linspace(start = first_range, stop = y_max, num = 5)
```

```python
colors = ['blue','cornflowerblue','gold','orange','red']
```

```python
plt.figure(figsize=(12,9))
plt.scatter(x, y, c=label, cmap=matplotlib.colors.ListedColormap(colors))
plt.ylabel('Temperature,°C')
plt.xlabel('Date')

plt.show()
```
